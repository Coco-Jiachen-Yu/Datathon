```python
msa_list = df['MSA ABBREVIATED NAME'].tolist()
```

```python
msa_sim = []
for x in msa_list:
    first = x.split(',')[0].lower()
    state = x.split(',')[1]
    state = state[1:3]
    first = first.split('-')[0]
    first = first.split('/')[0]
    msa_sim.append([first, state])
```

```python
msa_sim
```

```python
shared_msa = []
for i in data.items():
    name = i[1].split(',')[0].lower()
    name = name.split('-')[0]
    st = i[1].split(',')[1][1:3]
    for n in msa_sim:
        if name == n[0] and st == n[1]:
            shared_msa.append(i[0])
        
```
